```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import re
import matplotlib.pyplot as plt
import calendar

# load the data
df=pd.read_csv(" NOAA_datas.csv", error_bad_lines=False)
df=df.sort_values(by='Date')

#convert to celsius
df['Data_Value']=df['Data_Value']/10

# select 2015 data
df_15=df.loc[df['Date'].str.startswith('2015')].copy()

# select data before 2015
values=df_15['Date'].unique()
df_05_14=df[df.Date.isin(values)==False].copy()

#remove years
df_05_14['Date']=df_05_14['Date'].str.replace('^[0-9]*-','',regex=True)
df_15['Date']=df_15['Date'].str.replace('^[0-9]*-','',regex=True)

# remove leap day
df_05_14.drop(df_05_14.loc[df_05_14['Date']=='02-29'].index,inplace=True)


# filter data to either max or min and groupby month-day
max_05_14 = df_05_14.groupby(['Date']).agg({'Data_Value': max}).reset_index().rename(columns={'Data_Value': 'Max_in_05_14'})
min_05_14 = df_05_14.groupby(['Date']).agg({'Data_Value': min}).reset_index().rename(columns={'Data_Value': 'Min_in_05_14'})
max_15 = df_15.groupby(['Date']).agg({'Data_Value': max}).reset_index().rename(columns={'Data_Value': 'Max_in_15'})
min_15 = df_15.groupby(['Date']).agg({'Data_Value': min}).reset_index().rename(columns={'Data_Value': 'Min_in_15'})



# select max and min values from 2015 that are greater or lower than the recorded max in 2005-2014
greater_15=max_05_14.merge(max_15,on='Date')
lower_15=min_05_14.merge(min_15,on='Date')
higher_in_15=greater_15[greater_15.Max_in_15 >greater_15.Max_in_05_14].copy()
higher_in_15.drop('Max_in_05_14', axis=1, inplace=True)
lower_in_15=lower_15[lower_15.Min_in_05_14 > lower_15.Min_in_15].copy()
lower_in_15.drop('Min_in_05_14', axis=1, inplace=True)


# plot the min and max lines
plt.figure(figsize=(20,10))
plt.plot(max_05_14['Max_in_05_14'],color='tab:red',label='Max temperature recorded each day: 2005-2014')
plt.plot(min_05_14['Min_in_05_14'],color='tab:blue',label='Min temperature recorded each day: 2005-2014')
plt.legend()


# add the fill, between min and max
plt.fill_between(max_05_14.index, max_05_14.Max_in_05_14, min_05_14.Min_in_05_14, alpha=0.10, color='tab:grey')

# add points greater than max or less than min

plt.scatter(higher_in_15.index, higher_in_15.Max_in_15,s=100, c='purple',label='New max temperature in 2015')
plt.scatter(lower_in_15.index, lower_in_15.Min_in_15, s=100, c='orange',label='New min temperature in 2015')
plt.legend()

# set plot xlim
plt.xlim(-5, 370)

# tick locations
ticks=[-5, 0, 31, 59, 90, 120, 151, 181, 212, 243, 273, 304, 334, 365, 370]

# tick labels
labels = list(calendar.month_abbr)  # list of months
labels.extend(['Jan', ''])

# add the custom ticks and labels
plt.xticks(ticks=ticks, labels=labels)

# plot cosmetics
plt.xlabel('Day of Year: 0-365 Displaying Start of Month')
plt.ylabel('Temperature C')
plt.title('Daily Max and Min: 2005 - 2014\nRecorded Max < 2015 Temperatures < Recorded Min')
#plt.tight_layout()

plt.savefig('plot.jpg', dpi=300, bbox_inches='tight')
```
